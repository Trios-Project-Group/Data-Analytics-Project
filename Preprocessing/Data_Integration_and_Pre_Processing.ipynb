```python
#Importing required packages
import pandas as pd
import numpy as np
```

# Data Preprocessing and Feature Engineering

```python
df1=pd.read_csv('Dataset/Development.csv') #Dataset1 - development courses related to software
df1
```

```python
df2=pd.read_csv('Dataset/IT_Software.csv')#dataset2 - courses related to It and Software.
df2.drop(['is_wishlisted', 'discount_price__price_string','price_detail__price_string'], axis=1, inplace=True)
df2
```

```python
df=pd.concat([df1, df2], ignore_index=True, verify_integrity=True) #Conactenate two datasets - dataset1 and dataset2
df
```

```python
#Columns of the dataset
df.columns
```

```python
#Columns which have Nan Values
for i in range(len(df.columns)):
    rows=[]
    if df[df.columns[i]].isnull().values.any():
        print(df.columns[i])
        #print([k for k,j in enumerate(df[df.columns[i]].isnull()) if j])
        print(df[df.columns[i]].isnull().sum())
```

<h3>Replacing Null Values with useful information in dataset</h3>

```python
#Replace these null values with appropriate data
df['discount_price__currency'].value_counts() #All are INR hence all Nan values also would be INR
df['discount_price__currency'].fillna('INR',inplace=True)
df['discount_price__currency'].value_counts()
```

```python
df['price_detail__currency'].value_counts() #All are INR hence all Nan values also would be INR
df['price_detail__currency'].fillna('INR',inplace=True)
df['price_detail__currency'].value_counts()
```

```python
df['price_detail__amount'].count() #Those that are Nan are missed as their records are not available or free in this case we won't take it as free but 
df['price_detail__amount'].interpolate(inplace=True) #Average of upper and lower values in same column
df['price_detail__amount'].count()
```

```python
#Replacing Null values in discount with original price as no discount is given on those courses.
print(df['discount_price__amount'].isnull().values.any())
for i,j in enumerate(df['discount_price__amount'].isnull()):
    if j:
        df['discount_price__amount'][i] = df['price_detail__amount'][i]
print(df['discount_price__amount'].isnull().values.any())        
```

```python
#Check if Null values are present.
df.isnull().values.any()
```

<h3>Removing duplicates from merged dataset</h3>

```python
#Check all those indexes having duplicate records due to merging.
l = []
for i,j in enumerate(df['id'].value_counts()):
    if j==2:
        l.append(df['id'].value_counts().index[i])
print(l)
```

```python
#Total count of records before removing duplicates.
df.count()
for i in l:
    df.drop(df[df['id']==i].index[0], inplace=True)
df.count()
```

```python
l = []
for i,j in enumerate(df['id'].value_counts()):
    if j==2:
        l.append(df['id'].value_counts().index[i])
print(l)#Empty list implies duplicates are removed
```
